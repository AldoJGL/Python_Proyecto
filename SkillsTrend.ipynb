#Cuales son las habilidades mas demandadas para los analistas de datos?

Metodologia

1. Agregar el conteo mensual de habilidades.
2. Re-analizar basado en el porcentaje total de los trabajos.
3. Graficar la demanda mensual de habilidades

```python
#Importar Librerias
import ast
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Cargar Datos
df = pd.read_csv("hf://datasets/lukebarousse/data_jobs/data_jobs.csv")

#Limpiar Datos
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

columnas_ordenadas = df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index
df_DA_US_pivot = df_DA_US_pivot[columnas_ordenadas]

df_DA_US_pivot
```

```python
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)

df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent = df_DA_US_percent[df_DA_US_percent['job_posted_month_no'] != 'Total'].copy()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')

df_DA_US_percent
```

```python
df_plot = df_DA_US_percent.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Top habilidades para Analistas de datos en E.E.U.U')
plt.ylabel('Tendencia en las ofertas de empleo')
plt.xlabel('2024')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
  plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])

plt.show()
```
